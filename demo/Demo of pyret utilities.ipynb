# Demo of pyret utilities
This demo showcases some of the useful features of `pyret`, a library for analyzing retinal data.

In this demo, we will:

1. Load data from a retinal experiment
2. Bin spikes and estimate a firing rate
3. Compute receptive fields using the spike triggered average
4. Visualize receptive fields

```python
# Setup: we'll import the numpy and matplotlib packages, and use inline graphics
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### 1. Load experiment data

We are going to load data from a white noise experiment, in which 11 cells were recorded in response to spatiotemporal binary white noise. The stimulus has been downsampled to (50x50) spatial checkers with time points sampled every 10ms for 1000 seconds (roughly 16 minutes).

The spatiotemporal stimulus is stored in the `stimulus` variable, with a corresponding time vector stored in the `time` variable. Spike times for each of 11 cells in response to this stimulus are stored as arrays in a list called `spikes`.

```python
# Load experiment data
data = np.load('sample_data.npz')
data.keys()
```

### 2. Bin spikes and estimate a firing rate
One common task people might want to do with spiking data is to bin in into a firing rate. Here, we show how to do this easily using `pyret`:

```python
# import the pyret module for dealing with spike trains
import pyret.spiketools as spkt

# bin spike times with a single trial and 100ms bins.
numTrials = 1
binsize   = 0.01

# we use the binspikes function to bin our spike counts
bspk, tbins = spkt.binspikes(data['spikes'], binsize=binsize)

# now let's estimate a firing rate (smoothed spike counts) from our binned spikes
sigma = 0.02 # width of our Gaussian filter
rate = spkt.estfr(tbins, bspk, sigma)

# plot the spike count and firing rate for the first 10 seconds
fig = plt.figure(figsize=(12,6))
sns.set_style('whitegrid')
ax = fig.add_subplot(111);
ax.stem(tbins[:100], bspk[:100]/binsize, linefmt='b-', markerfmt='bo', basefmt='');
ax.plot(tbins[:100], rate[:100], 'r-');
ax.legend(('Firing Rate', 'Binned Spikes'), 'upper left', fontsize=20, fancybox=True, frameon=True);
ax.set_xlabel('Time (s)', fontsize=20);
ax.set_ylabel('Rate (Hz)', fontsize=20);
ax.set_xticklabels(ax.get_xticks(), fontsize=18);
ax.set_yticklabels(ax.get_yticks(), fontsize=18);
ax.set_ylim([0,135])
```

### 3. Compute receptive fields
Now we will compute a spatiotemporal receptive field (RF) using the spike-triggered average (STA). This is just a one line operation using `pyret`!

```python
# import the filter tools module
import pyret.filtertools as ft

# how long in the past to compute the filter for (time steps)
history = 10

# compute the STA
sta, tax = ft.getsta(data['time'], data['stimulus'], data['spikes'], history)
```

### 4. Visualize receptive fields
Our STA is now a (13 x 13 x 10) dimensional numpy array. Fortunately, `pyret` has some quick and easy ways to visualize receptive fields.

```python
# load visualizations module
import pyret.visualizations as viz

# find the peak (center) of the STA, and the spatial frame at which this peak is most clearly seen
center, timeslice = ft.filterpeak(sta)[1:];

# plot the spatial receptive field
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111)
viz.spatial(sta[:,:,timeslice], ax);

# plot the temporal receptive field
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.set_title('Temporal Filter')
ax = viz.temporal(tax,sta[center[1], center[0], :], ax);
```
